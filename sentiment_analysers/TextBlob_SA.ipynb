_*TextBlob is a higher-level library built on NLTK that provides a simple API for sentiment analysis._

```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
# Read CSV

data_file = pd.read_csv("Datafiniti_Hotel_Reviews.csv")
```

```python
# Convert ratings from float to integer

data_file['reviews.rating'] = data_file['reviews.rating'].astype(int)

# Convert text to string

data_file['reviews.text'] = data_file['reviews.text'].astype(str)
```

```python
# Simplify the dataframe

data = data_file[['id', 'reviews.rating', 'reviews.text']]
```

```python
data
```

```python
from textblob import TextBlob

# Generate the 5-star classification with TextBlob

def sentiment_TextBlob(text):
    """
    Analyzes text strings and determines the overall polarity of the text.
    It classifies between 1 to 5 stars based on the polarity score.
    """
    blob = TextBlob(text)
    polarity_score = blob.sentiment.polarity

    if polarity_score >= 0.6:
        return 5  # Very positive
    elif polarity_score >= 0.2:
        return 4  # Positive
    elif polarity_score >= -0.2:
        return 3  # Neutral
    elif polarity_score >= -0.6:
        return 2  # Negative
    else:
        return 1  # Very negative
```

```python
import time

# Process the reviews
print("Starting sentiment analysis...")
start_time = time.time()

# Add the TextBlob Sentiment to the dataframe
data['textblob.sentiment'] = data['reviews.text'].apply(lambda x: sentiment_TextBlob(x) if x != "nan" else '')  # Filters out observations with no text


print(f"Processing completed in {(time.time() - start_time) / 60:.2f} minutes")
```
